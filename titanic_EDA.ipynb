```python
import pandas as pd
titanic_df = pd.read_csv('titanic.csv')
titanic_df.head()
titanic_df.info()
titanic_df.describe()

titanic_df.isnull().sum()

titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)
titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0], inplace=True)
titanic_df.drop('Cabin', axis=1, inplace=True)

titanic_df['Pclass'] = titanic_df['Pclass'].astype('category')
titanic_df['Survived'] = titanic_df['Survived'].astype('category')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.histplot(titanic_df['Age'], kde=True)
plt.title('Age Distribution')
plt.show()


plt.figure(figsize=(6, 4))
sns.countplot(data=titanic_df, x='Survived')
plt.title('Survival Count')
plt.show()
```

```python

plt.figure(figsize=(8, 6))
sns.countplot(data=titanic_df, x='Pclass', hue='Survived')
plt.title('Survival Count by Passenger Class')
plt.show()

# relationship between 'Survived' and 'Age'
plt.figure(figsize=(10, 6))
sns.kdeplot(data=titanic_df, x='Age', hue='Survived', fill=True)
plt.title('Survival Distribution by Age')
plt.show()
```

```python
# relationship between 'Pclass', 'Sex', and 'Survived'
plt.figure(figsize=(10, 6))
sns.catplot(data=titanic_df, x='Pclass', hue='Survived', col='Sex', kind='count')
plt.show()

#  relationship between 'Fare', 'Pclass', and 'Survived'
plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic_df, x='Pclass', y='Fare', hue='Survived')
plt.title('Fare Distribution by Class and Survival')
plt.show()
```
